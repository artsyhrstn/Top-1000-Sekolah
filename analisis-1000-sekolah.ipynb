```python
# Download dataset
path = kagglehub.dataset_download("rezkyyayang/top-1000-sekolah-utbk-ltmpt")

# Direktori penyimpanan file

destination_path = "data"
if not os.path.exists(destination_path):
    os.makedirs(destination_path)

for file_name in os.listdir(path):
    full_file_name = os.path.join(path, file_name)
    if os.path.isfile(full_file_name):
        shutil.copy(full_file_name, destination_path)

print("Path to dataset files:", path)
print("Files saved to:", destination_path)
```

```python
geocode_data = pd.read_json('data/seribu_sekolah_geocode.json')
geocode_data['Sekolah'] = geocode_data['Sekolah'].apply(lambda x: x.split(',')[0])
geocode_data
```

```python
# Normalize geocode data into separate columns for latitude and longitude
geocode_df = pd.DataFrame(geocode_data)
geocode_df[['lat', 'lng']] = geocode_df.apply(lambda row: pd.Series({'lat': row['lat'], 'lng': row['lng']}), axis=1)

geocode_df
```

```python
# Perform a left join
merged_df = seribu_sekolah_df.merge(geocode_df, how='left', left_on='Sekolah_KabKota_Provinsi', right_on='Sekolah')

# Display the merged DataFrame
merged_df.to_csv('data/merged.csv', index=False)
```

```python
merged_df.sort_values(by='lat', inplace=True)
missing_lat_count = merged_df['lat'].isna().sum()
missing_lat_count
merged_df = pd.read_csv('data/merged.csv')
merged_df= merged_df.sort_values(by='lat', ascending=False)
merged_df.to_csv('data/merged.csv', index=False)
merged_df
```

```python
# Peta di Indonesia
m = folium.Map(location=[-2.548926, 118.0148634], zoom_start=5)

# Penanda di setiap sekolah
for index, row in merged_df.iterrows():
    if not pd.isna(row['lat']) and not pd.isna(row['lng']):
        folium.Circle(
            location=[row['lat'], row['lng']],
            radius=100,
            popup=row['Sekolah'],
            color='blue',
            fill=True,
            fill_color='blue'
        ).add_to(m)

# Tampilkan Peta
m
```
